```python
# Mengimpor LabelEncoder dari library sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder

# Membuat objek LabelEncoder untuk setiap kolom kategorikal
le_sex = LabelEncoder() # Untuk kolom 'Sex'
le_resting_ecg = LabelEncoder() # Untuk kolom 'RestingECG'
le_exercise_angina = LabelEncoder() # Untuk kolom 'ExerciseAngina'
le_st_slope = LabelEncoder() # Untuk kolom 'ST_Slope'

# Mengubah data kategorikal menjadi numerik menggunakan Label Encoding
dataset['Sex'] = le_sex.fit_transform(dataset['Sex']) # Mengubah kolom 'Sex'
dataset['ChestPainType'] = le_sex.fit_transform(dataset['ChestPainType']) # Mengubah kolom 'ChestPainType'
dataset['RestingECG'] = le_resting_ecg.fit_transform(dataset['RestingECG']) # Mengubah kolom 'RestingECG'
dataset['ExerciseAngina'] = le_exercise_angina.fit_transform(dataset['ExerciseAngina']) # Mengubah kolom 'ExerciseAngina'
dataset['ST_Slope'] = le_st_slope.fit_transform(dataset['ST_Slope']) # Mengubah kolom 'ST_Slope'
```

```python
# Variabel `x` menyimpan fitur-fitur (input) untuk model machine learning.
# `dataset.drop()` menghapus kolom 'HeartDisease' dari DataFrame `dataset`.
# `axis=1` menunjukkan bahwa yang dihapus adalah kolom.
# Jadi, `x` berisi semua kolom kecuali 'HeartDisease'.
x = dataset.drop(['HeartDisease'], axis=1)
y = dataset['HeartDisease']
#Variabel `y` menyimpan target (output) yang akan diprediksi oleh model.
# `dataset['HeartDisease']` memilih kolom 'HeartDisease' dari DataFrame `dataset`.
# Jadi, `y` berisi hanya nilai-nilai dari kolom 'HeartDisease'.
```

```python
from sklearn.model_selection import train_test_split
# Mengimpor fungsi `train_test_split` dari modul `sklearn.model_selection`.
# Fungsi ini digunakan untuk membagi data menjadi data latih dan data uji.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=8)
# Membagi data fitur (`x`) dan target (`y`) menjadi data latih dan data uji
```

```python
# Mengimpor kelas SVC (Support Vector Classifier) dari modul sklearn.svm.
# SVC digunakan untuk membangun model klasifikasi Support Vector Machine.
from sklearn.svm import SVC

# Membuat objek model SVC dengan nama 'model'.
model = SVC()
# Melatih model SVC menggunakan data latih (x_train dan y_train).
model.fit(x_train, y_train)
```

```python
# Membuat array NumPy bernama 'seseorang' yang berisi data satu individu.
# Data tersebut merepresentasikan fitur-fitur (input) untuk model machine learning.
# Fitur-fitur tersebut kemungkinan adalah: Age, Sex, ChestPainType, RestingBP, Cholesterol, FastingBS, RestingECG, MaxHR, ExerciseAngina, Oldpeak, ST_Slope
seseorang = np.array([[57, 0, 1, 130, 236, 0, 0, 174, 0, 0.0, 1]])

# Memprediksi status penyakit jantung ('HeartDisease') untuk 'seseorang' menggunakan model yang telah dilatih.
print(model.predict(seseorang))
```

```python
# Mengimpor fungsi accuracy_score dari library sklearn.metrics
from sklearn.metrics import accuracy_score

# Memprediksi target (HeartDisease) untuk data latih (x_train) menggunakan model yang sudah dilatih
y_pred = model.predict(x_train)
# Mencetak akurasi model pada data latih.
print(accuracy_score(y_train, y_pred))

# Memprediksi target (HeartDisease) untuk data uji (x_test) menggunakan model yang sudah dilatih.
y_pred_test = model.predict(x_test)
# Mencetak akurasi model pada data uji.
print(accuracy_score(y_test, y_pred_test))
```

```python
# Mengimpor GridSearchCV dari sklearn.model_selection
from sklearn.model_selection import GridSearchCV

# Mendefinisikan parameter yang akan dicoba oleh GridSearchCV.
# 'C', 'gamma', dan 'kernel' adalah parameter untuk model SVC.
params = {'C': [0.1, 1, 10, 100],
              'gamma': [1, 0.1, 0.01, 0.001],
              'kernel': ['rbf', 'linear']}

# Membuat objek GridSearchCV dengan model SVC, parameter yang akan diuji
grid = GridSearchCV(SVC(), params, refit=True, verbose=3)
# Melatih GridSearchCV dengan data latih (x_train dan y_train).
grid.fit(x_train, y_train)
```

```python
# mencetak skor terbaik yang dicapai oleh model selama proses Grid Search.
print(grid.best_score_)
```

```python
# Baris ini mengambil parameter terbaik yang ditemukan oleh GridSearchCV.
# - `grid.best_estimator_`: Mengacu pada model terbaik yang ditemukan selama proses pencarian.
best_params = grid.best_estimator_.get_params()

# Loop ini akan mengiterasi setiap parameter yang didefinisikan di `params`
# Baris ini mencetak nama parameter (`param`) dan nilai terbaiknya (`best_params[param]`) yang ditemukan oleh GridSearchCV.
for param in params:
    print(f"{param} : {best_params[param]}")
```
